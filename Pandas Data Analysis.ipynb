```python
import pandas as pd
import numpy as np
```

```python
#reading in all csv files

departments=pd.read_csv("departments.csv")
dept_emp=pd.read_csv("dept_emp.csv")
dept_manager=pd.read_csv("dept_manager.csv")
employees=pd.read_csv("employees.csv")
salaries=pd.read_csv("salaries.csv")
titles=pd.read_csv("titles.csv")
```

```python
#ordering employee df so it lines up with salaries df
employees=employees.sort_values(by=['emp_no'],ascending=True)
employees
```

```python
salaries=salaries.sort_values(by=['emp_no'],ascending=True)
salaries.head()
```

```python
#-- Question 1: List the employee number, last name, first name, sex, and salary of each employee.
q1_df = pd.DataFrame(columns=['Employee #','Last Name','First Name','Sex','Salary'])
q1_df['Employee #']=employees['emp_no']
q1_df['Last Name']=employees['last_name']
q1_df['First Name']=employees['first_name']
q1_df['Sex']=employees['sex']
q1_df['Salary']=salaries['salary']
q1_df=q1_df.sort_values(by=['Employee #'],ascending=True)
q1_df=q1_df.reset_index()
q1_df=q1_df.drop(columns='index')
q1_df.head()
```

```python
#Question2: List the first name, last name, and hire date for the employees who were hired in 1986. Ordered by hire date ascending for readability.
q2_df = pd.DataFrame(columns=['First Name','Last Name','Hire Date'])
q2_df['First Name']=employees['first_name']
q2_df['Last Name']=employees['last_name']

q2_df['Hire Date']=employees['hire_date']
q2_df['Hire Date']=pd.to_datetime(q2_df['Hire Date'])
q2_df=q2_df.loc[(q2_df["Hire Date"]>='1986-01-01') & (q2_df["Hire Date"]<='1986-12-31')]
q2_df=q2_df.sort_values(by='Hire Date', ascending=True)                   
q2_df.head()
```

```python
# Question 3: List the manager of each department along with their department number, department name, employee number, last name, and first name.
#creating the dataframe
q3_df = pd.DataFrame(columns=['dept_num','dept_name','emp_no','last_name','first_name'])
```

```python
#getting the department name for each manager
dept=[]
for index,row in dept_manager.iterrows():
    if row['dept_no']=='d001':
        dept.append('Marketing')
    elif row['dept_no']=='d002':
        dept.append('Finance')
    elif row['dept_no']=='d003':
        dept.append('Human Resources')
    elif row['dept_no']=='d004':
        dept.append('Production')
    elif row['dept_no']=='d005':
        dept.append('Development')
    elif row['dept_no']=='d006':
        dept.append('Quality Management')
    elif row['dept_no']=='d007':
        dept.append('Sales')
    elif row['dept_no']=='d008':
        dept.append('Research')
    elif row['dept_no']=='d009':
        dept.append('Customer Service')
```

```python
#pulling manger first name and last name and putting in dataframe
list_num = dept_manager['emp_no'].tolist()
l_name=[]
f_name=[]
for index,row in employees.iterrows():
    if row['emp_no'] in list_num:
        l_name.append(row['last_name'])
        f_name.append(row['first_name'])
```

```python
#populating the rest of the dataframe
q3_df = pd.DataFrame(columns=['dept_num','dept_name','emp_no','last_name','first_name'])
q3_df['dept_num']=dept_manager['dept_no']
q3_df['dept_name']=dept
q3_df['emp_no']=dept_manager['emp_no']
q3_df['last_name']=l_name
q3_df['first_name']=f_name
```

```python
#Question 4:List the department number for each employee along with that employeeâ€™s employee number, last name, first name, and department name.
#pulling in department numbers
emp_num = pd.unique(employees['emp_no'].tolist())

dept_num=[]
for index,row in dept_emp.iterrows():
    if (row['emp_no'] in emp_num) & (row['emp_no']<=499999):
        dept_num.append(row['dept_no'])




```

```python
#pulling in department names
dept_name1 = pd.unique(dept_emp['emp_no'].tolist())
dept_name=[]
for index,row in dept_emp.iterrows():
    if (row['dept_no']=='d001') & (row['emp_no'] in dept_name1):
        dept_name.append('Marketing')
    elif (row['dept_no']=='d002') & (row['emp_no'] in dept_name1):
        dept_name.append('Finance')
    elif (row['dept_no']=='d003') & (row['emp_no'] in dept_name1):
        dept_name.append('Human Resources')
    elif (row['dept_no']=='d004') & (row['emp_no'] in dept_name1):
        dept_name.append('Production')
    elif (row['dept_no']=='d005') & (row['emp_no'] in dept_name1):
        dept_name.append('Development')
    elif (row['dept_no']=='d006') & (row['emp_no'] in dept_name1):
        dept_name.append('Quality Management')
    elif (row['dept_no']=='d007') & (row['emp_no'] in dept_name1):
        dept_name.append('Sales')
    elif (row['dept_no']=='d008') & (row['emp_no'] in dept_name1):
        dept_name.append('Research')
    elif (row['dept_no']=='d009') & (row['emp_no'] in dept_name1):
        dept_name.append('Customer Service')
    else:
        pass
```

```python
#Question 4:List the department number for each employee along with that employeeâ€™s employee number, last name, first name, and department name.
#creating DF
q4_df=pd.DataFrame(columns=['dept_num','emp_no','last_name','first_name','dept_name'])
#populating DF
q4_df['dept_num']=dept_num
q4_df['emp_no']=employees['emp_no']
q4_df['last_name']=employees['last_name']
q4_df['first_name']=employees['first_name']
q4_df['dept_name']=dept_name


```

```python
#dropping null values and presenting final DF
q4_df=q4_df.dropna()
q4_df
```

```python
#Question 7: List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name.
    
#pulling all employees from DF that are in sales or development depts
sales_dev=[]
for index,row in dept_emp.iterrows():
    if (row['dept_no'] == 'd007') | (row['dept_no'] == 'd005'):
        sales_dev.append(row['emp_no'])    
    
```

```python
#adding raw department name to DF
q7_df['dept_name']=dept_number
```

```python
#Getting the actual department name from our department numbers
dept_num_list = q7_df['dept_name'].tolist()
dept_name2=[]
for row in dept_num_list:
    if row=='d005':
        dept_name2.append('Development')
    elif row=='d007':
        dept_name2.append('Sales')
    else:
        pass
```
